```python
import pandas as pd
import numpy as np

import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize

from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
from sklearn.linear_model import LogisticRegression

#nltk.download('punkt')
#nltk.download('stopwords')
```

```python
df = pd.read_csv("Data/mail_data.csv") 
```

```python
df['Category'] = df['Category'].map({'ham':0, 'spam':1})

x = df['Message']
y = df['Category']
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    x, y, test_size=0.2, random_state=42
)
```

```python
from sklearn.feature_extraction.text import CountVectorizer

bow = CountVectorizer(
    stop_words='english',
    max_features=3000
)

X_train_bow = bow.fit_transform(X_train)
X_test_bow = bow.transform(X_test)
```

```python
model_bow = LogisticRegression(max_iter=1000)
model_bow.fit(X_train_bow, y_train)
```

```python
y_pred_bow = model_bow.predict(X_test_bow)

print("BoW Accuracy:", accuracy_score(y_test, y_pred_bow))
print(confusion_matrix(y_test, y_pred_bow))
print(classification_report(y_test, y_pred_bow))
```

```python
new_mail = ["Congratulations! You won a lottery. Claim now."]  # list of new mails
```

```python
# Step 1: Transform new mail using the trained BoW vectorizer
new_bow = bow.transform(new_mail)

# Step 2: Predict
pred_bow = model_bow.predict(new_bow)

# Step 3: Print result
print("BoW Prediction:", "Spam" if pred_bow[0]==1 else "Ham")
```

### Word2-vec

```python
def tokenize(text):
    tokens = word_tokenize(text.lower())
    tokens = [w for w in tokens if w.isalpha()]
    tokens = [w for w in tokens if w not in stopwords.words('english')]
    return tokens

X_train_tokens = X_train.apply(tokenize)
X_test_tokens = X_test.apply(tokenize)
```

```python
from gensim.models import Word2Vec

w2v_model = Word2Vec(
    sentences=X_train_tokens,
    vector_size=100,
    window=5,
    min_count=2,
    workers=4
)
```

```python
def vectorize_sentence(tokens, model):
    vectors = [model.wv[word] for word in tokens if word in model.wv]
    if len(vectors) == 0:
        return np.zeros(model.vector_size)
    return np.mean(vectors, axis=0)

X_train_w2v = np.array([vectorize_sentence(s, w2v_model) for s in X_train_tokens])
X_test_w2v = np.array([vectorize_sentence(s, w2v_model) for s in X_test_tokens])
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_train_w2v = scaler.fit_transform(X_train_w2v)
X_test_w2v = scaler.transform(X_test_w2v)
```

```python
model_w2v = LogisticRegression(max_iter=1000)
model_w2v.fit(X_train_w2v, y_train)
```

```python
y_pred_w2v = model_w2v.predict(X_test_w2v)

print("Word2Vec Accuracy:", accuracy_score(y_test, y_pred_w2v))
print(confusion_matrix(y_test, y_pred_w2v))
print(classification_report(y_test, y_pred_w2v))
```

```python
# Step 1: Tokenize
new_tokens = [tokenize(mail) for mail in new_mail]

# Step 2: Convert to average Word2Vec vector
new_w2v = np.array([vectorize_sentence(tokens, w2v_model) for tokens in new_tokens])

# Step 3: Scale
new_w2v = scaler.transform(new_w2v)

# Step 4: Predict
pred_w2v = model_w2v.predict(new_w2v)

# Step 5: Print
print("Word2Vec Prediction:", "Spam" if pred_w2v[0]==1 else "Ham")
```
